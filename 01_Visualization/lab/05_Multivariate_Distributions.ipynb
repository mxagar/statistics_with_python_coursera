# Multivariate Distributions

This notebook contains an example that shows how in multivariate datasets variables need to be plotted jointly in scatterplots to check for relationships that are not visible if plotted inseparate distirbutions.

Take-Home message: **Always plot sctterplots and histograms of pairs of variables!**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
# We define a multivariate (bivariate) dataset with two normal distributons
# They have different meas
# BUT: theirs covariance matrix contains a 
# Recall: covariance matrix:
# cov(X) = [var(x_i,x_j)]
# var(x_i,x_j) = sum((x_i - mean(x_i))^2) * sum((x_j - mean(x_j))^2)
# Here: var(x,y) = r, which controls how the relationship is
# r = 1: positive linear
# r = -1: negative linear
r = 1
mean = [15, 5]
cov = [[1, r], [r, 1]]
x, y = np.random.multivariate_normal(mean, cov, 400).T
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(x = x, bins = 15)
plt.title("X")

plt.subplot(1,2,2)
plt.hist(x = y, bins = 15)
plt.title("Y")

plt.show()
```

```python
# VERY IMPORTANT:
# Even though we seem to have separate normal distributions
# a joint scatterplot shows a perfect linear relationship!
plt.figure(figsize=(10,10))
plt.subplot(2,2,2)
plt.scatter(x = x, y = y)
plt.title("Joint Distribution of X and Y")

plt.subplot(2,2,4)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")


plt.subplot(2,2,1)
plt.hist(x = y, orientation = "horizontal", bins = 15)
plt.title("Marginal Distribution of Y")

plt.show()
```
